#MIN MAX & Z SCORE NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
df=pd.DataFrame({'X':[4,8,13,7],'Y':[11,4,5,14]})
scaler=MinMaxScaler(feature_range=(10,20))
a=scaler.fit_transform(df)
print(df)
print(a)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() #same as Z-score
sdf=scaler.fit_transform(df)
print(sdf)
```

#PRINCIPAL COMPONENT ANALYSIS

```python
#WITHOUT NORMALIZATION
from sklearn.decomposition import PCA
pca=PCA(n_components=1)
principalComponents=pca.fit_transform(df)
print(principalComponents)
```

```python
pdf=pd.DataFrame(data=principalComponents,columns=['PC1'])
pdf
```

```python
#WITH NORMALIZATION
scaler=MinMaxScaler()
a=scaler.fit_transform(df)
pca=PCA(n_components=1)
p=pca.fit_transform(a)
print(p)
```

#IRIS DATASET

```python
data=pd.read_csv('/content/iris_data.csv')
```

```python
data
```

```python
data1=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
data1
```

```python
S=StandardScaler()
```

```python
std_data=S.fit_transform(data1)
```

```python
std_data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(std_data)
print(principalComponents)
pdf=pd.DataFrame(data=principalComponents,columns=['PC1','PC2'])
pdf
```

```python
import matplotlib.pyplot as plt
plt.scatter(pdf['PC1'],pdf['PC2'])
```
